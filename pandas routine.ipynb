```python
r = range(1971,2015)
map(str, r)
cols = ['Country Name','Country Code'] + list(map(str, r))
cols[:5]
```

```python
df_agri.loc[:,cols].melt(id_vars=['Country Name','Country Code']).rename(columns={'variable':'date','Country Name':'Country','Country Code':'name','value':'agriculture'})
```

```python
def fun_formato(df, col = 'agriculture'):
    df = df.loc[:,cols].melt(id_vars=['Country Name','Country Code']).rename(columns={'variable':'date','Country Name':'Country','Country Code':'name','value':col})
    df['date'] = df['date'].astype(float)
    return df
```
