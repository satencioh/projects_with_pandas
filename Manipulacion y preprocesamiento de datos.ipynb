### Estructuras de dataframes en detalle

```python
df_meteoritos.nunique()
```

```python
#convertimos a fall y nametyoe en variables categoricas
df_meteoritos[['fall','nametype']] = df_meteoritos[['fall','nametype']].astype('category')
```

```python
df_meteoritos.dtypes
```

```python
#para conocer los dos tipos de categoria de la variable fall
df_meteoritos['fall'].unique()
```

```python
#para conocer la frecuencia en la que aparecen Found y Fell
df_meteoritos['fall'].value_counts()
```

```python
#podemos convertir los valores de las variables categoricas en 
#variables dummy
pd.get_dummies(df_meteoritos['fall'])
```

```python
#para agregar una nueva columna a nuestro df
df_meteoritos['ones'] = 1
df_meteoritos
```

```python
#agregamos las variables dummies a dos nuevas columnas
df_meteoritos[['Fell','Found']] = pd.get_dummies(df_meteoritos['fall'])
df_meteoritos
```

```python
df_meteoritos['year']
```

```python
#convertimos la columna year en fecha
pd.to_datetime(
    df_meteoritos['year'], 
    errors='coerce', 
    format='%m/%d/%Y %H:%M:%S %p'
)
```

```python
#y sobreescribimos year
df_meteoritos['year'] = pd.to_datetime(
    df_meteoritos['year'], 
    errors='coerce', 
    format='%m/%d/%Y %H:%M:%S %p'
)
df_meteoritos
```

```python
#la nueva conversion de variables
df_meteoritos.convert_dtypes().dtypes
```

```python
#renonbramos la columna 'mass (g)' por 'mass', 
#inplace para indicar que queremos que se guarden los cambios directamente en el mismo df
df_meteoritos.rename(columns={'mass (g)':'mass'}, inplace=True)
```
