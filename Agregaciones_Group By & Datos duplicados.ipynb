```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('diamonds')
df
```

<b>Agregaciones_Group By</b>

```python
df.groupby('cut').mean()
```

```python
df.groupby('cut').median()
```

```python
df.groupby('cut')['price'].mean()
```

```python
for key_group, group in df.groupby('cut'):
    grouped_price = group['price'].mean()
    print('Cut: {}, Price: {}\n'.format(key_group,grouped_price))
```

```python
df.groupby(['cut','color'])['price'].mean().to_frame()
```

```python
df.groupby(['cut','color'])['price'].aggregate(['min',np.mean,max])
```

```python
def mean_kilo(x):
    return np.mean(x)/1000
```

```python
df.groupby(['cut','color'])['price'].aggregate(['min',np.mean,max, mean_kilo]).head(5)
```

```python
dict_agg = {'carat':[min,max],'price':[np.mean,mean_kilo]}
dict_agg
```

```python
df.groupby(['cut','color']).aggregate(dict_agg).head(5)
```

```python
def f_filter(x):
    return mean_kilo(x['price'])>4
```

```python
df.groupby('cut').filter(f_filter)
```

```python
df.groupby('cut').filter(f_filter)['cut'].unique()
```

<b>Datos Duplicados</b>

```python
df_d = pd.DataFrame({'a': ['w'] * 4 + ['x'] * 3 + ['y'] * 2 + ['z']+['v'], 
                   'b': [1, 1, 1, 1, 2, 2, 2, 3, 3, 4,5]})
df_d
```

```python
#para encontrar los registros duplicados (los que se marcan en True cumplen)
df_d.duplicated()
```

```python
#para identificar los casos que no tiene duplicados (cRemarco el hecho de que usÃ© negaciÃ³n '~' para ver los registros no duplicados.)
df_d[~ df_d.duplicated()]
```

```python
#si me interesara ver cuÃ¡les son los registros duplicados, podemos usar keep=False
df_d[df_d.duplicated(keep=False)]
```

```python
#puedes usar el comando 'drop_duplicates' para eliminar los duplicados. Por defecto, la funciÃ³n guarda el primer resultado keep='first'
df_d.drop_duplicates()
```

```python
#si quieres solo borrar duplicados teniendo en cuenta una sola columna, lo puedes hacer mediante una lista nombrando las columnas donde vas a eliminar los duplicados, en este caso, ['a']
df_d.drop_duplicates(['a'],keep='last')
```
